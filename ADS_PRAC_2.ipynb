```python
import pandas as pd
```

```python
df=pd.read_csv('Bengaluru_House_Data.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#filling null values with previous value
df1=df.fillna(method='pad')
df1
```

```python
#filling null values
df2=df.fillna(value=0)
df2
```

```python
#filling null value with next value
df3=df.fillna(method='bfill')
df3
```

```python
#filling null values with previous column value
df4=df.fillna(method='pad',axis=1)
df4
```

```python
#filling null values with next volumn value
df5=df.fillna(method='bfill',axis=1)
df5
```

```python
#filling diff values in null in diff columns
df6=df.fillna({'society': 'abcd',
               'balcony': 'defg'})
df6
```

```python
#filling null values with mean of column
df7=df.fillna(value=df['balcony'].mean())
df7
```

Dropping null value column

```python
df8=df.dropna()
df8
```

```python
df9=df.dropna(how='all')
df9
```

```python
df10=df.dropna(how='any')
df10
```

```python
#replace
import numpy as np
df11=df.replace(to_replace=np.nan,value=8456)
df11
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

```python
#interpolate()
df['balcony']=df['balcony'].interpolate(method='linear')
df
```
